**1. Write  a computer program capable of reducing the number of intensity levels  in an image from 256 to 2, in integer powers of 2. The desired number of  intensity levels needs to be a variable input to your program.**

```python
def reduce_intensity(image, num_levels):
    return (image // (256/num_levels)) * (256/num_levels)
```

```python
# number of intensity levels 256->128->64->32->16->8->4->2
plt.axis('off')
plt.imshow(reduce_intensity(img_gray,256), cmap='gray');
```

**3 .Rotate the image by 45 and 90 degrees (Matlab provides simple command lines for doing this).**

```python
rows, cols, channels = img_rgb.shape

#Get rotate matrix M, center of image, rotate = 45, scale = 1
M = cv.getRotationMatrix2D((cols/2,rows/2),45,1)
#Rotate image with dsize = (rows,cols)
dst_45 = cv.warpAffine(img_rgb,M,(cols,rows))

M = cv.getRotationMatrix2D((cols/2,rows/2),90,1)
#Rotate image with dsize = (rows,cols)
dst_90 = cv.warpAffine(img_rgb,M,(cols,rows))

f, rotate = plt.subplots(1,2,figsize=(10,30))
rotate[0].imshow(dst_45);
rotate[1].imshow(dst_90);
rotate[0].set_title('45 degree')
rotate[1].set_title('90 degree')
rotate[0].axis('off');
rotate[1].axis('off');
```

**4. For every 33 block of the image (without overlapping), replace all corresponding 9 pixels by their average. This operation simulates reducing the image spatial resolution. Repeat this for 55 blocks and 77 blocks. If you are using Matlab, investigate simple command lines to do this important operation.**

```python
#Block size
stride = (3, 5, 7, 19)
rows, cols, channels = img_rgb.shape

#Init subplot with 1 row, len(stride) columns, figure size = (50,100)
f, axarr = plt.subplots(1, len(stride),figsize=(50,100))

for i,S in enumerate(stride):
    im_mod = img_rgb.copy()
    #loop over row and col of pixel with step S
    for r in range(0,rows,S):
        for c in range(0, cols, S):
            for ch in range(0, channels):
                #calculate the average pixel value
                im_mod[r:r+S,c:c+S,ch]=np.average(img_rgb[r:r+S,c:c+S,ch])
    axarr[i].imshow(im_mod);
    axarr[i].set_title('{}x{}'.format(S,S),fontsize=50);
    axarr[i].axis('off');                    
```
